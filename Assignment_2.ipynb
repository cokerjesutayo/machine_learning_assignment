# Section 3. Feature Scaling

Feature scaling is the process of bringing all data values to the same range so no feature is too big or too small.

Why is it necessary?
Feature scaling makes all values similar in size so no feature dominates because without it large values can mislead the model.


The model would focus too much on features with large values and ignore the smaller ones leading to poor and inaccurate results.

## 10. Min-Max

```python
cols_to_scale=["housing_median_age","total_rooms","median_income","median_house_value"]
```

```python
for col in cols_to_scale:
  col_min=df[col].min() #find the min value in the column
  col_max=df[col].max() #find the max value in the column
  df[col+"_scaled"]=(df[col]- col_min)/ (col_max-col_min) #Apply the formula to every row in the column
```

```python
#check that every scaled column is now between 0 and 1
for col in cols_to_scale:
  scaled_col=col+"_scaled"
  print(f"{scaled_col} : min {df[scaled_col].min():.4f} max {df[scaled_col].max():.4f}")
```

```python
print(df[[col, col+"_scaled"]].head(10))
```

The scaled values are just the original values made smaller so they all fit between 0 and 1 so  when the original number increases, the scaled one also increases only the size changes
